```python
import pandas as pd 
import os

result = pd.DataFrame(columns=['time','Open','High','Low','Close'])
tickers = [f for f in os.listdir() if os.path.isfile(f) and f.split('.')[-1]=='csv']
for ticker in tickers:
    try:
        df = pd.read_csv(f'{ticker}', header=None, usecols=[0,1,2,3,4],names=['time','Open','High','Low','Close'])
        df['ticker'] = ticker.split('.')[0]
        result = pd.concat([result, df])
    except:
        print(ticker)
        break

from datetime import datetime, timedelta
result = result.iloc[1:,:]
def clean(d):
    d = str(d)[0:-6]
    return d

t = result['time'].apply(clean)
result['time'] = pd.to_datetime(t)

result.set_index('time')


```

```python
import os

curr = pd.Timestamp(year=2022, month=10, day=19)
end = datetime(year=2022, month=11, day=11)

while not curr == end:
    curr_str = curr.strftime('%Y-%m-%d')
    today = result[result['time'].dt.date.astype(str) == curr_str]
    path = os.path.dirname(os.path.abspath('split.ipynb'))+'/filtered/'+curr.strftime('%Y-%m-%d'+'.csv')
    today.to_csv(path)
    curr += pd.Timedelta(days = 1)
```
